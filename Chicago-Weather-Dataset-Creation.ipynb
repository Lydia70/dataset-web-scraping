# Chicago Weather Dataset

## Dreaming of a White Christmas?

In this notebook, I show you how I create a nice dataset including Chicago weather data at Christmas day. This dataset is available at Kaggle. I use web scraping to retrieve the data from the website. Then, I do some data cleaning, and the dataset is ready for data analysis.

```python
#import packages and libraries
#import packages
import pandas as pd
import csv

import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as plt
import seaborn as sns

import requests # library to handle requests
import random # library for random number generation

from bs4 import BeautifulSoup #for webscraping

from html_table_extractor.extractor import Extractor

#other packages
#!pip install selenium
from selenium import webdriver
```

```python
#connect with the website
url = 'https://www.weather.gov/lot/Chicago_Christmas_Data'
dfs = pd.read_html(url)

print(len(dfs))
```

```python
#get the table
df = dfs[1]
df.columns = df.iloc[0]
df = df.iloc[1: , :]
df
```

```python
#Add columns
df['White Christmas'] = ""
```

```python
#Replace values in columns
df['Snow'] = df['Snow'].replace("M","")
df['Snow'] = df['Snow'].replace("T","")
```

```python
df.head(10)
```

```python
df.tail(10)
```

```python
#fill column white christmas
Christmas = []

for row in df['Snow']:
    if row != "" and row != 0:
        Christmas.append(bool(1))
    elif row == 0:
        Christmas.append(bool(0))
    else:
        Christmas.append("Not Defined")

df['White Christmas'] = Christmas
```

```python
df.head(20)
```

```python
#Replace values in columns
df['High Temp'] = df['High Temp'].replace("M","")
df['Min Temp'] = df['Min Temp'].replace("M","")
df['Precip'] = df['Precip'].replace("T","")
```

```python
#Rename columns
df = df.rename(columns={'High Temp':'Fahrenheit High Temp'})
df = df.rename(columns={'Min Temp':'Fahrenheit Min Temp'})
df = df.rename(columns={'Precip':'Precipitation'})
```

```python
df['Celsius Min Temp'] = ""
df['Celsius High Temp'] = ""
```

```python
df['Fahrenheit Min Temp'] = df['Fahrenheit Min Temp'].replace("",500)
df['Fahrenheit Min Temp'] = df['Fahrenheit Min Temp'].astype(int)
```

```python
#calculate minimum temperature in degree Celsius 
df['Celsius Min Temp'] = (df['Fahrenheit Min Temp']-32)*0.5556 
```

```python
df['Fahrenheit High Temp'] = df['Fahrenheit High Temp'].replace("",500)
df['Fahrenheit High Temp'] = df['Fahrenheit High Temp'].astype(int)
```

```python
#calculate high temperature in degree Celsius
```

```python
df['Celsius High Temp'] = (df['Fahrenheit High Temp']-32)*0.5556 
```

```python
#replace missing values again
df['Fahrenheit Min Temp'] = df['Fahrenheit Min Temp'].replace(500,"")
df['Fahrenheit High Temp'] = df['Fahrenheit High Temp'].replace(500,"")
```

```python
#replace incorrect values
df['Celsius Min Temp'] = df['Celsius Min Temp'].replace(260.0208,"")
df['Celsius High Temp'] = df['Celsius High Temp'].replace(260.0208,"")
```

```python
#Export to CSV file
df.to_csv('...../ChicagoWeatherChristmas.csv')
```
